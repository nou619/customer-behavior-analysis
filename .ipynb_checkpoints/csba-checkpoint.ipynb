```python
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.isnull().sum()
```

```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns
```

```python
df["age"].unique()
```

```python
df["gender"].unique()
```

```python
df["payment method"].unique()
```

```python
labels=['young adult','adult','middle aged','senior']
df['age group']=pd.qcut(df['age'],q=4,labels=labels)#doesnt work if the age is between 70 and 100 for example, there's no young adult aged 70 yo lol
```

```python
df[['age','age group']].head(40)
```

```python
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90}
df['purchase frequency days']=df['frequency of purchases'].map(frequency_mapping)
```

```python
df[['purchase frequency days','frequency of purchases']].head(20)
```
